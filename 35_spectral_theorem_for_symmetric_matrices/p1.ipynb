```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.patches import Ellipse
```

```python
f: lambda x1, x2: (5/8)*x1**2 - 6/8*x1*x2 + (5/8)*x2**2
A = np.array([[5 / 8, -3 / 8], [-3 / 8, 5 / 8]])
```

```python
lamdbas, V = np.linalg.eigh(A)
print(lamdbas)
print(V)
```

```python
semi_axes = 1/np.sqrt(lamdbas)
v1, v2 = V
print(semi_axes)
print(v1)
print(v2)
```

```python
fig, ax = plt.subplots()

# Set the center of the ellipse
center = (0, 0)

# Set the angle of rotation
angle = np.degrees(np.arctan2(v2[1], v2[0]))

# Create the ellipse
ellipse = Ellipse(
  center, width=2 * semi_axes[0], height=2 * semi_axes[1], angle=angle, fill=False
)

# Add the ellipse to the plot
ax.add_patch(ellipse)

# Plot the eigenvectors
ax.quiver(0, 0, v1[0], v1[1], angles='xy', scale_units='xy', scale=1, color='pink', label='v1')
ax.quiver(0, 0, v2[0], v2[1], angles='xy', scale_units='xy', scale=1, color='orange', label='v2')

# Set the aspect ratio of the plot
ax.set_aspect("equal")

# Set the x and y limits of the plot
ax.set_xlim(-2, 2)
ax.set_ylim(-2, 2)

# Show the grid
ax.grid(True)

# Show the legend
ax.legend()

# Show the plot
plt.show()
```
